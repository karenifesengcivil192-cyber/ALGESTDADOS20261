```python
dia = int(input("Número do dia (1-7): "))

match dia:
    case 1:
        print("Segunda-feira")
    case 2:
        print("Terça-feira")
    case 3:
        print("Quarta-feira")
    case 4:
        print("Quinta-feira")
    case 5:
        print("Sexta-feira")
    case 6 | 7:  # o | significa 'ou'
        print("Fim de semana!")
    case _:  # _ é o 'else' do match
        print("Dia inválido")
```
